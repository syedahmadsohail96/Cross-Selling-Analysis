```python
# Define the salary ranges
salary_ranges = {
    'Under 40K': (0, 40000),
    '40K-60K': (40000, 60000),
    '60K-80K': (60000, 80000),
    '80K-100K': (80000, 100000),
    '100K-120K': (100000, 120000),
    '120K-140K': (120000, 140000),
    '140K-160K': (140000, 160000),
    '160K-180K': (160000, 180000),
    'Above 180K': (180000, float('inf'))
}

# Define the order of the salary ranges for the x-axis labels
ordered_salary_ranges = [
    'Under 40K',
    '40K-60K',
    '60K-80K',
    '80K-100K',
    '100K-120K',
    '120K-140K',
    '140K-160K',
    '160K-180K',
    'Above 180K'
]

# Define a function to map salary ranges to labels
def get_salary_range_label(salary):
    for label, (lower, upper) in salary_ranges.items():
        if lower <= salary < upper:
            return label
    return 'Unknown'

# Apply the function to the Gross_Income column and create a new column 'Salary_Range'
merged_df['Salary_Range'] = merged_df['Gross_Income'].apply(get_salary_range_label)

# Filter the data to only include individuals with Home_Account > 0
home_account_filter = merged_df['Home_Account'] > 0
home_account_df = merged_df[home_account_filter]

# Group the data by Salary_Range and count the number of rows in each group
salary_counts = home_account_df.groupby('Salary_Range').size()

# Reorder the salary counts based on the order of the x-axis labels
salary_counts = salary_counts.reindex(ordered_salary_ranges)

# Create a bar chart of the salary ranges and the number of rows in each range
plt.bar(salary_counts.index, salary_counts.values)

# Add axis labels and a title
plt.xlabel('Salary Range')
plt.ylabel('Number of Home Accounts')
plt.title('Home Accounts by Salary Range')

# Rotate the x-axis labels to avoid overlapping
plt.xticks(rotation=45, ha='right')

# Display the chart
plt.show()
```

# Cross Selling Analysis with Payroll:


```python
# Filter the data for individuals with Payroll > 0 and Mortgage < 0 and Pensions < 0
cross_sell_filter = (merged_df['Payroll'] > 0) & (merged_df['Mortgage'] > 0) 
cross_sell_df = merged_df[cross_sell_filter]

# Count the number of individuals with Payroll > 0 and Mortgage < 0 and Pensions < 0
cross_sell_count = len(cross_sell_df)

# Count the number of individuals without Pensions or Mortgage and with Payroll > 0
no_cross_sell_filter = (merged_df['Payroll'] > 0) & (merged_df['Mortgage'] == 0) 
no_cross_sell_df = merged_df[no_cross_sell_filter]

# Count the number of individuals without Pensions or Mortgage and with Payroll > 0
no_cross_sell_count = len(no_cross_sell_df)

# Create a pie chart
labels = ['Cross-sell', 'No cross-sell']
sizes = [cross_sell_count, no_cross_sell_count]
colors = ['orange', 'blue']
explode = (0.1, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)

# Add title
plt.title('Payroll with Mortgage Cross-selling')

# Show the chart
plt.show()
```
